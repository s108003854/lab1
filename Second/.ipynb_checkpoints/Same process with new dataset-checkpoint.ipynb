```python
# import newdataset 

import pandas as pd
import helpers.data_mining_helpers as dmh

imdb_data, imdb_score = dmh.load_data('imdb_labelled.txt')
imdb_category_name = ['imdb' for i in range(len(imdb_data))]
imdb_category = [0 for i in range(len(imdb_data))]

amazon_data, amazon_score = dmh.load_data('amazon_cells_labelled.txt')
amazon_category_name = ['amazon' for i in range(len(amazon_data))]
amazon_category = [1 for i in range(len(amazon_data))]

yelp_data, yelp_score = dmh.load_data('yelp_labelled.txt')
yelp_category_name = ['yelp' for i in range(len(yelp_data))]
yelp_category = [2 for i in range(len(yelp_data))]

data = imdb_data + amazon_data + yelp_data
score = imdb_score + amazon_score + yelp_score
category_name = imdb_category_name + amazon_category_name + yelp_category_name
category = imdb_category + amazon_category + yelp_category
```

```python
#Converting into Pandas Dataframe
X = pd.DataFrame.from_records(data, columns= ['text'])

#Add column
X['score'] = score
X['category_name'] = category_name
X['category'] = category

# a simple query
X[0:10][["text", "category_name"]]
```

```python
#check missing value
X.isnull()
# dummy record as dictionary format
dummy_dict = [{'text': 'dummy_record','category': 1}]
X = X.append(dummy_dict, ignore_index=True)
print("add dummy data:"+str(len(X)))
X.isnull().apply(lambda x: dmh.check_missing_values(x))
X.dropna(inplace=True)
X.isnull().apply(lambda x: dmh.check_missing_values(x))
print("remove dummy data:"+str(len(X)))
```

```python
#Dealing with Duplicate Data
dummy_duplicate_dict = [{
                             'text': 'dummy record',
                             'category': 1, 
                             'category_name': "dummy category"
                        },
                        {
                             'text': 'dummy record',
                             'category': 1, 
                             'category_name': "dummy category"
                        }]
#add duplicate Data
X = X.append(dummy_duplicate_dict, ignore_index=True)
sum(X.duplicated('text'))
print("add duplicate data:"+str(len(X)))
X.drop_duplicates(keep=False, inplace=True) # inplace applies changes directly on our dataframe
print("remove duplicate data:"+str(len(X)))
```

```python
#Data Preprocessing
#Sampling
X_sample = X.sample(n=100) #random state
print("length of sample data: ",len(X_sample))
X_sample[0:4]
```

```python

import matplotlib.pyplot as plt
print(X_sample.category_name.value_counts())

# plot barchart for X_sample
X_sample.category_name.value_counts().plot(kind = 'bar',
                                    title = 'Category distribution',
                                    ylim = [0, 50],        
                                    rot = 0, fontsize = 11, figsize = (8,3))
```

```python
#Feature Creation
# get unigram token
X['unigrams'] = X['text'].apply(lambda x: dmh.tokenize_text(x))
X
```

```python
#Feature subset selection
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer()
# transorfmed the data into a term-document matrix
X_counts = count_vect.fit_transform(X.text)

# We can check the shape of this matrix by:
X_counts.shape
count_vect.get_feature_names()[0:10]
```

```python
#term-document matrix
# first twenty features only
plot_x = ["term_"+str(i) for i in count_vect.get_feature_names()[0:20]]
# obtain document index
plot_y = ["doc_"+ str(i) for i in list(X.index)[0:100]]
plot_z = X_counts[0:100, 0:20].toarray()


import seaborn as sns

df_todraw = pd.DataFrame(plot_z, columns = plot_x, index = plot_y)
plt.subplots(figsize=(9, 7))
ax = sns.heatmap(df_todraw,
                 cmap="PuRd",
                 vmin=0, vmax=1, annot=True)
```

```python
#Dimensionality Reduction
from sklearn.decomposition import PCA
X_reduced = PCA(n_components = 2).fit_transform(X_counts.toarray())
print("reduced shape: "+str(X_reduced.shape))
```

```python
#plot scatter
col = ['coral', 'blue', 'm']
categories = ['imdb','yelp','amazon']
# plot
fig = plt.figure(figsize = (25,10))
ax = fig.subplots()

for c, cate in zip(col, categories):
    xs = X_reduced[X['category_name'] == cate].T[0]
    ys = X_reduced[X['category_name'] == cate].T[1]
   
    ax.scatter(xs, ys, c = c, marker='o')

ax.grid(color='gray', linestyle=':', linewidth=2, alpha=0.2)
ax.set_xlabel('\nX Label')
ax.set_ylabel('\nY Label')

plt.show()
```

```python
#Atrribute Transformation / Aggregation
import numpy as np
# reduce the amount of terms you want to compute frequencies for
term_frequencies = []
for j in range(0,X_counts.shape[1]):
    term_frequencies.append(sum(X_counts[:,j].toarray()))
    
term_frequencies = np.asarray(X_counts.sum(axis=0))[0]

plt.subplots(figsize=(100, 10))
g = sns.barplot(x=count_vect.get_feature_names()[:300], 
            y=term_frequencies[:300])
g.set_xticklabels(count_vect.get_feature_names()[:300], rotation = 90);
```

```python
#long tail
index = np.argsort(-term_frequencies)

feature_name = np.asarray(count_vect.get_feature_names())

plt.subplots(figsize=(100, 10))
g = sns.barplot(x=feature_name[index[1:100]], 
            y=term_frequencies[index[1:100]])
```

```python
import math
term_frequencies_log = [math.log(i) for i in term_frequencies]
term_frequencies_log_arr = np.asarray(term_frequencies_log)

index_log = np.argsort(-term_frequencies_log_arr)

feature_name = np.asarray(count_vect.get_feature_names())

plt.subplots(figsize=(100, 10))
g = sns.barplot(x=feature_name[index_log[1:100]], 
            y=term_frequencies_log_arr[index_log[1:100]])
```

```python
#Discretization and Binarization

from sklearn import preprocessing, metrics, decomposition, pipeline, dummy
mlb = preprocessing.LabelBinarizer()
mlb.fit(X.category)
X['bin_category'] = mlb.transform(X['category']).tolist()
X
```

```python
#Data Exploration
document_to_transform_1 = []
random_record_1 = X.iloc[50]
random_record_1 = random_record_1['text']
document_to_transform_1.append(random_record_1)

document_to_transform_2 = []
random_record_2 = X.iloc[100]
random_record_2 = random_record_2['text']
document_to_transform_2.append(random_record_2)

document_to_transform_3 = []
random_record_3 = X.iloc[150]
random_record_3 = random_record_3['text']
document_to_transform_3.append(random_record_3)
print(document_to_transform_1)
print(document_to_transform_2)
print(document_to_transform_3)
```

```python
from sklearn.preprocessing import binarize

# Transform sentence with Vectorizers
document_vector_count_1 = count_vect.transform(document_to_transform_1)
document_vector_count_2 = count_vect.transform(document_to_transform_2)
document_vector_count_3 = count_vect.transform(document_to_transform_3)

# Binarize vecors to simplify: 0 for abscence, 1 for prescence
document_vector_count_1_bin = binarize(document_vector_count_1)
document_vector_count_2_bin = binarize(document_vector_count_2)
document_vector_count_3_bin = binarize(document_vector_count_3)

# print
print("Let's take a look at the count vectors:")
print(document_vector_count_1.todense())
print(document_vector_count_2.todense())
print(document_vector_count_3.todense())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
# Calculate Cosine Similarity
cos_sim_count_1_2 = cosine_similarity(document_vector_count_1, document_vector_count_2, dense_output=True)
cos_sim_count_1_3 = cosine_similarity(document_vector_count_1, document_vector_count_3, dense_output=True)
cos_sim_count_1_1 = cosine_similarity(document_vector_count_1, document_vector_count_1, dense_output=True)
cos_sim_count_2_2 = cosine_similarity(document_vector_count_2, document_vector_count_2, dense_output=True)

# Print 
print("Cosine Similarity using count bw 1 and 2: %(x)f" %{"x":cos_sim_count_1_2})
print("Cosine Similarity using count bw 1 and 3: %(x)f" %{"x":cos_sim_count_1_3})
print("Cosine Similarity using count bw 1 and 1: %(x)f" %{"x":cos_sim_count_1_1})
print("Cosine Similarity using count bw 2 and 2: %(x)f" %{"x":cos_sim_count_2_2})
```
